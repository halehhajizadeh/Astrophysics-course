```python
import numpy as np
import matplotlib.pylab as plt
```

```python
D = [['IRAS',0.57],['KAO',0.9],['SOFIA',2.6],['ISO',0.6]]
```

```python
#infrared
l_min = 7*1e-7
l_max = 1e-3
```

```python
l = 10**np.linspace(np.log10(l_min),np.log10(l_max),100)
```

```python
D[0]
```

```python
for i in range(4):
    theta = l/D[i][1]
    plt.plot(l,theta,label = str(D[i][0]))
plt.legend()
plt.xlabel('$\lambda (m)$',fontsize = 13)
plt.ylabel('$\Delta$($\u03F4$)',fontsize = 13)  
plt.savefig('dtheta-lambda')
```
