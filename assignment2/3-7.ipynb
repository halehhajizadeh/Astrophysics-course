```python
import numpy as np
import matplotlib.pylab as plt
```

```python
h = 6.63*10**(-27)
c = 3*10**8
k = 1.38*10**(-16)
```

```python
T = [35000,21000,13500,9700,8100,7200,6500,6000,5400,4700,4000,3300,2600]
```

```python
l = 10**np.linspace(-8 , -3 , 100)
```

```python
T = np.array(T)
```

```python
for i in range(0,13):
    log_I = (2*h*c**2)/((l**5)*(np.exp(h*c/(l*k*T[i]))-1))
    plt.plot(l,log_I, label = str(T[i]))
    plt.xlim(0,0.000001)
plt.legend()
plt.xlabel('$\lambda$(m)',fontsize = 13)
plt.ylabel('I($\lambda$,T)(erg/s)',fontsize = 13)
plt.savefig('I-lambda')
```

```python
for i in range(len(T)):
    log_I = np.log10((2*h*c**2)/((l**5)*(np.exp(h*c/(l*k*T[i]))-1)))
    plt.plot(l,log_I, label = str(T[i]))
    plt.xlim(0,0.00000021)
plt.legend()
plt.xlabel('$\lambda$(m)',fontsize = 13)
plt.ylabel('log I($\lambda$,T)(erg/s)',fontsize = 13)
plt.savefig('logI-lambda')
```
